```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
survival_data = titanic_df['survived'].value_counts().sort_index()
plt.figure(figsize=(6,4))
sns.barplot(x=survival_data.index, y=survival_data.values, palette='Blues_d')
plt.title('Overall Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Passenger Count')
plt.grid(True)
plt.show()
```

```python
gender_counts = titanic_df['sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops={'width':0.4})
plt.title('Passenger Gender Ratio')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(titanic_df['age'], bins=30, kde=True, color='coral')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
pclass_counts = titanic_df['pclass'].value_counts().sort_index()
plt.figure(figsize=(6,4))
sns.barplot(x=pclass_counts.values, y=['1st', '2nd', '3rd'], palette='mako')
plt.title('Passenger Count by Class')
plt.xlabel('Number of Passengers')
plt.ylabel('Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='sex', y='survived', data=titanic_df, ci=None, palette='spring')
plt.title('Average Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='pclass', y='survived', data=titanic_df, ci=None, palette='autumn')
plt.title('Average Survival Rate by Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=titanic_df[titanic_df['survived'] == 1], x='pclass', hue='sex', palette='Set2')
plt.title('Survivors by Gender and Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survivor Count')
plt.show()
```

```python
embark_counts = titanic_df['embarked'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=embark_counts.index, y=embark_counts.values, palette='cool')
plt.title('Passengers by Embarkation Point')
plt.xlabel('Port')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='embarked', y='survived', data=titanic_df, ci=None, palette='viridis')
plt.title('Average Survival Rate by Embarkation Port')
plt.ylabel('Survival Rate')
plt.show()
```

```python
titanic_df['family_members'] = titanic_df['sibsp'] + titanic_df['parch']
plt.figure(figsize=(8,4))
sns.countplot(x='family_members', hue='survived', data=titanic_df, palette='cubehelix')
plt.title('Survival by Family Size')
plt.xlabel('Family Members Aboard')
plt.ylabel('Passenger Count')
plt.legend(title='Survived')
plt.show()
```

```python
titanic_df['fare_group'] = pd.cut(titanic_df['fare'], bins=[0, 20, 50, 100, 600], labels=['Low', 'Med', 'High', 'Very High'])

plt.figure(figsize=(6,4))
sns.countplot(x='fare_group', hue='survived', data=titanic_df, palette='Spectral')
plt.title('Survival by Fare Group')
plt.xlabel('Fare Range')
plt.ylabel('Passenger Count')
plt.legend(title='Survived')
plt.show()
```

```python
titanic_df['family_size_group'] = pd.cut(titanic_df['family_members'], bins=[-1, 0, 2, 5, 10], labels=['Solo', 'Small', 'Medium', 'Large'])
age_filtered = titanic_df.dropna(subset=['age', 'family_size_group'])
plt.figure(figsize=(10,6))
sns.boxplot(data=age_filtered, x='family_size_group', y='age', hue='sex', palette='Set3')
plt.title('Age Distribution by Family Size Group and Gender')
plt.xlabel('Family Size Group')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```
